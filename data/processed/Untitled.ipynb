```python
import pandas as pd

# Load the dataset
df_path = "rahvad.csv"
df = pd.read_csv(df_path)

# Function to clean column names
def clean_column_name(name):
    # Remove leading ".." and capitalize the first letter
    name = name.lstrip('.').strip()  # Remove leading dots and any extra spaces
    return name.capitalize()

# Apply the function to each column name
df.columns = [clean_column_name(col) for col in df.columns]

# Save the updated DataFrame to a new CSV file
new_df_path = "rahvad_cleaned.csv"
df.to_csv(new_df_path, index=False)

print(f"Updated DataFrame saved to {new_df_path}")
```

```python
import pandas as pd

# Load the dataset
df_path = "rahvad_cleaned.csv"  # Use the cleaned CSV if available
df = pd.read_csv(df_path)

# Define the exact locations to filter
exact_locations_to_filter = ['..Tallinn', '..Tartu linn']

# Remove rows where 'Maakond' is exactly '..Tallinn' or '..Tartu linn'
df_filtered = df[~df['Maakond'].isin(exact_locations_to_filter)]

# Create a new DataFrame df_tt with only '..Tallinn' and '..Tartu linn'
df_tt = df[df['Maakond'].isin(exact_locations_to_filter)]

# Save the filtered DataFrame to a new CSV file
filtered_df_path = "rahvad_filtered.csv"
df_filtered.to_csv(filtered_df_path, index=False)

# Save the new DataFrame df_tt to a separate CSV file
df_tt_path = "rahvad_tt.csv"
df_tt.to_csv(df_tt_path, index=False)

print(f"Filtered DataFrame saved to {filtered_df_path}")
print(f"DataFrame with Tallinn and Tartu saved to {df_tt_path}")
```
